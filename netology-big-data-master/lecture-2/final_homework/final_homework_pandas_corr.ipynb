```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('ml-wine').getOrCreate()
```

```python
df = spark.read.csv('winequality-white.csv', header = True, inferSchema = True, sep=";")
```

```python
import pandas as pd
pd.DataFrame(df.take(5), columns=df.columns).transpose()
```

```python
numeric_features = [t[0] for t in df.dtypes if t[1] == 'double']
df.select(numeric_features).describe().toPandas().transpose()
```

```python
numeric_data = df.select(numeric_features).toPandas()
axs = pd.plotting.scatter_matrix(numeric_data, figsize=(16, 16));
n = len(numeric_data.columns)
for i in range(n):
    v = axs[i, 0]
    v.yaxis.label.set_rotation(0)
    v.yaxis.label.set_ha('right')
    v.set_yticks(())
    h = axs[n-1, i]
    h.xaxis.label.set_rotation(90)
    h.set_xticks(())
```
