```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
joined = ratings.merge(movies, on='movieId', how='left')
joined.head()
```

```python
(joined[['movieId','title', 'rating']]
.query('rating == 5.0')
.groupby(['movieId', 'title'])
.agg('count')
.sort_values(by='rating', ascending=False).head(1))
```

```python
#Задание 2 По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония)
# категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.
```

```python
power = pd.read_csv('power.csv')
power.head()
```

```python
(
    power[
        (power.year >= 2005) & 
        (power.year <= 2010) & 
        (power.category.isin([4, 12, 21])) & 
        (power.country.isin(['Latvia','Estonia','Lithuania']) &
        (power.quantity > 0)
         
        )
    ].quantity.sum()
)
```
